# Example of a function that is once but not differentiable

$$
f(x) = 
\begin{cases}
-\frac{1}{2} x^2 & \text{if $x < 0$}\\
\phantom{-}\frac{1}{2} x^2 & \text{otherwise}
\end{cases}
$$

```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## Define and plot the function

```python
def f(x):
    if x < 0:
        return -0.5 * x**2
    else:
        return 0.5 * x**2
f = np.vectorize(f)
```

```python
x = np.linspace(-1, 1, 1000)
y = f(x)
```

```python
sns.set_style('darkgrid')
plt.plot(x, y)
```

## Plot the first derivative

```python
h = 0.001
dy = (f(x + h) - f(x)) / h
```

```python
sns.set_style('darkgrid')
plt.plot(x, dy)
```

## Plot the second derivative

https://en.wikipedia.org/wiki/Finite_difference#Higher-order_differences

```python
h = 0.001
ddy = (f(x + h) - 2*f(x) + f(x - h)) / h**2
```

```python
sns.set_style('darkgrid')
plt.plot(x, ddy)
```

The discontinuation point at 0 means that $f$ is not twice differentiable.
